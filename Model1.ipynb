#**Predictions on test.csv (Target attribute: outage_type)**

#**Importing Libraries**

```
import pandas as pd
```

#**Reading Dataset**

```
df=pd.read_csv('/content/Broadband_data.csv')
del df['Unnamed: 0']
df.head(2)
```

```
df1=df.copy()
del df1['id']
df1.head(2)
```

```
df1.shape
```

**Converting Outage_type from Categorical to Nuemerical**

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df1.outage_type)
le.classes_
df1['outage_type']=le.transform(df1.outage_type)
```

```
df1.head(2)
```

**Onehot encoding**

```
new_df=pd.get_dummies(data=df1, columns=['area_code', 'broadband_type','log_report_type','transit_server_type'])
```

```
new_df.head(2)
```

**Training Set**

```
train=new_df.iloc[0:40000,:]
train=train.reset_index()
del train['index']
train.head()
```

```
train.shape
```

**Testing Set**

```
test=new_df.iloc[40000:]
test=test.reset_index()
del test['index']
test.head()
```

```
test.shape
```

**Preparing X and y in training set**

```
y=train.outage_type
y
```

```
X=train
del X['outage_type']
X.head()
```

**Train and validation split**

```
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.35, random_state=5)
```

# **Model Prediction**

**Logistic Regression**

```
 from sklearn.linear_model import LogisticRegression
```

```
 clf = LogisticRegression(multi_class='ovr').fit(X_train,y_train)
```

```
pred=clf.predict(X_val)
```

```
clf.predict_proba(X_val)
```

```
clf.score(X_train,y_train) # Accuracy on training set
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_val,pred)
```

```
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
print ('Accuracy Score :',accuracy_score(y_val, pred) ) # Validation Score
print ('Report : ')
print (classification_report(y_val, pred))
```

**Prediction on Testing set**

```
Ytest=test.outage_type # 
Ytest
```

```
xtest=test
del xtest['outage_type']
xtest.head()
```

```
pred_test_lr=clf.predict(xtest)
```

```
pred_test_lr
```

**Testing Score**

```
print ('Accuracy Score :',accuracy_score(Ytest, pred_test_lr) )  # Testing Score
```

```
print ('Report : ')
print (classification_report(Ytest,pred_test_lr))
```
